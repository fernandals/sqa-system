```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme(style="whitegrid", palette="pastel")
```

## Audio Quality Metric (UTMOS)

```python
title = "Audio Quality"
metric = "UTMOS"
score = "Quality"
```

```python
df = pd.read_csv(metric + "_comparison.csv")
#df = df.drop(columns=[metric.lower() + "_ground_truth"])
```

```python
# convert dataframe to right format
df_violin = df.melt(id_vars=["filename"], var_name="Model", value_name="Score")

plt.figure(figsize=(12, 7))
ax = sns.violinplot(
    x="Model", 
    y="Score", 
    data=df_violin, 
    inner=None,
    linewidth=1.5,
    saturation=0.75
)

plt.title(title + " Score Distribution by Model", fontsize=16, pad=20)
plt.xlabel("Model", fontsize=12, labelpad=10)
plt.ylabel(score + " Score", fontsize=12, labelpad=10)

plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.grid(True, axis='y', linestyle='--', alpha=0.7)

# ground truth line
plt.axhline(y=5.0, color='red', linestyle='--', linewidth=1.5, label="Ground Truth")
plt.legend(loc='upper left', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Convert dataframe to the right format
df_box = df.melt(id_vars=["filename"], var_name="Model", value_name="Score")

plt.figure(figsize=(12, 7))
ax = sns.boxplot(
    x="Model", 
    y="Score", 
    data=df_box, 
    linewidth=1.5,
    saturation=0.75
)

plt.title(title + " Score Distribution by Model", fontsize=16, pad=20)
plt.xlabel("Model", fontsize=12, labelpad=10)
plt.ylabel(score + " Score", fontsize=12, labelpad=10)

plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.grid(True, axis='y', linestyle='--', alpha=0.7)

# Ground truth line
plt.axhline(y=5.0, color='red', linestyle='--', linewidth=1.5, label="Ground Truth")
plt.legend(loc='upper left', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
df_lines = df.copy()

# assign numerical IDs instead of using filenames
df_lines["ID"] = range(1, len(df) + 1)

plt.figure(figsize=(18, 7))
colors = ["b", "g", "r", "c", "m", "y", "k"]

for idx, col in enumerate(df_lines.columns[1:-1]):
    plt.plot(
        df_lines["ID"], 
        df_lines[col], 
        marker="o", 
        linestyle="-", 
        label=col, 
        color=colors[idx % len(colors)],
        markersize=8,
        linewidth=2  
    )

plt.title(title + " Scores Across Audio Files", fontsize=16, pad=20)
plt.xlabel("Audio File ID", fontsize=12, labelpad=10)
plt.ylabel(score + " Score", fontsize=12, labelpad=10)

plt.legend(
    title="Model", 
    fontsize=10, 
    title_fontsize=14, 
    bbox_to_anchor=(1.05, 1),  # Move legend outside the plot
    loc="upper left"
)

plt.grid(True, linestyle="--", alpha=0.7) 
plt.xticks(df_lines["ID"], fontsize=10, rotation=60, ha="right")
plt.yticks(fontsize=10)

# ground truth line
plt.axhline(y=5.0, color="red", linestyle="--", linewidth=2, label="ground_truth")
plt.legend(loc='upper left', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
df_heatmap = df.copy()

# Set 'filename' as the index for the heatmap
df_heatmap.set_index("filename", inplace=True)

plt.figure(figsize=(16, 8))
sns.heatmap(df_heatmap, annot=False, cbar_kws={"label": f"{score} Score"})

plt.title(f"Heatmap of {score} Scores by File and Model", fontsize=16, pad=20)
plt.xlabel("Model", fontsize=12, labelpad=10)
plt.ylabel("Filename", fontsize=12, labelpad=10)

plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.tight_layout()
plt.show()
```
